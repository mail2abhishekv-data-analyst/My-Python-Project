```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt # visualizing data
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv(r'C:\Users\lappify\Desktop\Python\projects\Diwali Sales Data.csv', encoding= 'unicode_escape')
```

```python
df.info()
```

```python
df.drop(['Status','unnamed1'],axis=1,inplace=True)
```

```python
df.head()
```

```python
pd.isnull(df).sum()
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.shape
```

```python
df.dtypes
```

```python
df['Amount']=df['Amount'].astype(int)
```

```python
df.dtypes
```

```python
df.head()
```

```python
ax = sns.countplot(x = 'Gender', data = df)
for bars in ax.containers:
    ax.bar_label(bars) # 
```

```python
sales_gen=df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.barplot(x='Gender',y='Amount',data=sales_gen)
```

```python
ax = sns.countplot(x = 'Age Group', data = df, hue='Gender')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
df.columns
```

```python
sales_gen=df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.barplot(x='Age Group',y='Amount',data=sales_gen,hue='Amount',palette='Blues')
```

```python
state_order=df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(x='State',y='Orders',data=state_order,palette='viridis')
```

```python
state_order=df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(x='State',y='Amount',data=state_order,hue='Amount',palette='viridis')
```

```python
Maritul_status=df.groupby(['Marital_Status','Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)
sns.set(rc={'figure.figsize':(5,5)})
sns.barplot(x='Marital_Status',y='Amount',data=Maritul_status,hue='Gender', palette='bright')
```

```python
ax = sns.countplot(x='Marital_Status', data=df, hue='Marital_Status',palette='bright', legend=False)
sns.set(rc={'figure.figsize':(3,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sns.set(rc={'figure.figsize':(20,5)})
ax = sns.countplot(x = 'Occupation', data = df,hue='Occupation',palette='Greens')
for bars in ax.containers:
    ax.bar_label(bars) 
```

```python
Occupation=df.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(x='Occupation',y='Amount',data=Occupation,hue='Occupation', palette='Greens')
```

```python
sns.set(rc={'figure.figsize':(25,5)})
ax = sns.countplot(x = 'Product_Category', data = df,hue='Product_Category',palette='Blues')
for bars in ax.containers:
    ax.bar_label(bars) 
```

```python
Product_Category=df.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.set(rc={'figure.figsize':(25,5)})
sns.barplot(x='Product_Category',y='Amount',data=Product_Category,hue='Product_Category', palette='Blues')
```
